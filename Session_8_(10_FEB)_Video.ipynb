```python
import moviepy.editor as mp
import speech_recognition as sr
from textblob import TextBlob
import matplotlib.pyplot as plt

# Step 1: Extract Audio from the Video
def extract_audio(video_path, audio_path):
    video = mp.VideoFileClip(video_path)
    video.audio.write_audiofile(audio_path)

# Step 2: Transcribe Audio to Text
def transcribe_audio(audio_path):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = recognizer.record(source)

    try:
        text = recognizer.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        return "Could not understand audio"
    except sr.RequestError:
        return "Could not request results"

# Step 3: Analyze Emotion from Text using Sentiment Analysis
def analyze_emotion(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    # Emotion mapping based on polarity
    if sentiment > 0.1:
        return 'Positive'
    elif sentiment < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Step 4: Plot Flow of Emotions (Optional for sentence-based analysis)
def plot_emotion_flow(text):
    sentences = text.split('. ')
    emotions = [analyze_emotion(sentence) for sentence in sentences]

    # Count frequency of each emotion
    emotion_counts = {'Positive': emotions.count('Positive'),
                      'Negative': emotions.count('Negative'),
                      'Neutral': emotions.count('Neutral')}

    # Plotting the trend
    plt.bar(emotion_counts.keys(), emotion_counts.values())
    plt.xlabel('Emotion')
    plt.ylabel('Frequency')
    plt.title('Emotion Flow in the Video')
    plt.show()

# Main Function
def main(video_path):
    audio_path = 'audio.wav'
    extract_audio(video_path, audio_path)
    print("Audio extracted successfully.")

    text = transcribe_audio(audio_path)
    print("Transcription: ", text)

    emotion = analyze_emotion(text)
    print(f"Overall Emotion: {emotion}")

    plot_emotion_flow(text)

# Running the analysis on your video
video_path = "Colgate Strong Teeth _ Daant Strong Toh Main Strong _ English.mp4"
main(video_path)
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Extract audio and analyze pitch
def analyze_audio_pitch(audio_path):
    y, sr = librosa.load(audio_path)

    # Extract pitch and tempo
    pitches, magnitudes = librosa.core.piptrack(y=y, sr=sr)

    pitch_values = []
    for t in range(pitches.shape[1]):
        pitch = pitches[:, t]
        pitch_values.append(pitch[pitch > 0].mean())  # Use mean pitch for each frame

    # Plot pitch trend
    plt.plot(pitch_values)
    plt.xlabel('Frame')
    plt.ylabel('Pitch')
    plt.title('Audio Pitch Trend')
    plt.show()

# Running pitch analysis
audio_path = "audio.wav"  # Extracted from the video previously
analyze_audio_pitch(audio_path)
```

```python
from transformers import pipeline

# Step 1: Load pre-trained emotion recognition model
classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')

# Step 2: Analyze emotions in the transcribed text
def deep_emotion_analysis(text):
    possible_labels = ["positive", "negative", "neutral", "joy", "anger", "sadness", "fear", "surprise"]

    result = classifier(text, candidate_labels=possible_labels)
    return result['labels'][0], result['scores'][0]

# Example usage
text = "I am so happy to be part of this wonderful campaign!"
emotion, score = deep_emotion_analysis(text)
print(f"Detected Emotion: {emotion} with confidence: {score}")
```
